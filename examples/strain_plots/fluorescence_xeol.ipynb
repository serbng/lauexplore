# How to plot/access fluorescence and XEOL data using lauexplore and a H5 file

Use your own h5 file or the one available on https://drive.google.com/drive/folders/1LI0wUcVHP4Y87Q1txBCYmLaxa39rcPvv?usp=drive_link

```python
# Create the Scan object using the h5 file to have information about scan/map

#h5path = Path('path/to/file')
#scan  = Scan.from_h5(h5path, scan_number=2)


h5path = Path('F4_map1.h5')
scan  = Scan.from_h5(h5path, scan_number=2)
```

## Fluo object and plot using h5 file

```python
# Create the Strain object to plot the strain results using the Dataset object
fluo = Fluorescence.from_h5(h5path, scan_number=2, material='Ga')

# define your plot layoult
fig = fluo.plot(
    width = 1000, 
    height = 500, 
    colorscale = 'balance',
    zmin=0, 
    zmax = 25, 

    
)

fig.show()

```

## XEOL object and plot using h5 file

```python
xeol = XEOL.from_h5(h5path, 
                    scan_number=2, 
                    channel=200, 
                    roi = (500,600),
                    ref_range=(800,900), 
                    normalize_to_monitor=True)

xeol.plot(
    width=1000,
    height = 500)
```
